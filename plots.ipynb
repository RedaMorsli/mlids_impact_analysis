```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./output/DOS/16-02/data.csv")
attack_name = "DOS"
attack_abv = 'dos'
save_directory = "./output/DOS/16-02/plots/"

df.plot(x="timestamp", y="cpu_total")
plt.xlabel('Time')
plt.ylabel('CPU')
plt.title(attack_name + ' attack (CPU/Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_cpu_time.png")
plt.show()
```

```python
df.plot(x="timestamp", y="num_flows")
plt.xlabel('Time')
plt.ylabel('Number of flows')
plt.title(attack_name + ' attack (Number of flows/Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_num_flows_time.png")
plt.show()
```

```python
df.plot(x="timestamp", y="class")
plt.xlabel('Time')
plt.ylabel('Malicious flows')
plt.title(attack_name + ' attack (Malicious flows/Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_class_time.png")
plt.show()
```

```python
df.plot(x="timestamp", y="flow_duration")
plt.xlabel('Time')
plt.ylabel('Flow duration')
plt.title(attack_name + ' attack (Flow duration/Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_flow_duration_time.png")
plt.show()
```

```python
df.plot(x="timestamp", y="flow_byts_s")
plt.xlabel('Time')
plt.ylabel('Flow Bytes/s')
plt.title(attack_name + ' attack (Flow Bytes/s /Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_flow_byts_s_time.png")
plt.show()
```

```python
df.plot(x="timestamp", y="flow_pkts_s")
plt.xlabel('Time')
plt.ylabel('Flow Packets/s')
plt.title(attack_name + ' attack (Flow Packets/s /Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_flow_pkts_s_time.png")
plt.show()
```

```python
df.plot(x="timestamp", y=["tot_fwd_pkts", "tot_bwd_pkts"])
plt.xlabel('Time')
plt.ylabel('Total packets')
plt.title(attack_name + ' attack (Total packets/Time)')
plt.xticks([])
plt.savefig(save_directory + attack_abv + "_tot_pkts_time.png")
plt.show()
```
