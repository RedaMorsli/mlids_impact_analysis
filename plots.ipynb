```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("./output/SQLI-BFXSS/data1.csv")
df2 = pd.read_csv("./output/SQLI-BFXSS/data2.csv")
attack_name = "Web attacks"
attack_abv = 'web'
save_directory1 = "./output/SQLI-BFXSS/plots1/"
save_directory2 = "./output/SQLI-BFXSS/plots2/"
```

Method 1

```python
df1.plot(x="timestamp", y="cpu")
plt.xlabel('Time')
plt.ylabel('CPU')
plt.title(attack_name + ' attack (CPU/Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_cpu_time.png")
plt.show()
```

```python
df1.plot(x="timestamp", y="class")
plt.xlabel('Time')
plt.ylabel('Malicious flows')
plt.title(attack_name + ' attack (Malicious flows/Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_class_time.png")
plt.show()
```

```python
df1.plot(x="timestamp", y="flow_duration")
plt.xlabel('Time')
plt.ylabel('Flow duration')
plt.title(attack_name + ' attack (Flow duration/Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_flow_duration_time.png")
plt.show()
```

```python
df1.plot(x="timestamp", y="flow_byts_s")
plt.xlabel('Time')
plt.ylabel('Flow Bytes/s')
plt.title(attack_name + ' attack (Flow Bytes/s /Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_flow_byts_s_time.png")
plt.show()
```

```python
df1.plot(x="timestamp", y="flow_pkts_s")
plt.xlabel('Time')
plt.ylabel('Flow Packets/s')
plt.title(attack_name + ' attack (Flow Packets/s /Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_flow_pkts_s_time.png")
plt.show()
```

```python
df1.plot(x="timestamp", y=["tot_fwd_pkts", "tot_bwd_pkts"])
plt.xlabel('Time')
plt.ylabel('Total packets')
plt.title(attack_name + ' attack (Total packets/Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_tot_pkts_time.png")
plt.show()
```

```python
df1.plot(x="timestamp", y=["totlen_fwd_pkts", "totlen_bwd_pkts"])
plt.xlabel('Time')
plt.ylabel('Total length of packets')
plt.title(attack_name + ' attack (Total length of packets/Time)')
plt.xticks([])
plt.savefig(save_directory1 + attack_abv + "_totlen_pkts_time.png")
plt.show()
```
